```python

df_corr=df.corr()

plt.figure(figsize=(10,10))
sns.set(font_scale=0.8)
sns.heatmap(df_corr, annot=True, cbar=False)
plt.show()

```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x = df.drop('Rings',axis=1)
x.head()
```

```python
y = df['Rings']
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(x,
                                                  y,
                                                  test_size=0.2,
                                                  shuffle=True,
                                                  random_state=12)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(x_train, y_train)
ly_preds = model.predict(x_test)

print('평균제곱근오차', mean_squared_error(ly_preds, y_test))

def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

print('평균제곱근오차', mse_np(ly_preds, y_test))

def mse(actual, predicted):
    sum_square_error = sum((a - p) ** 2 for a, p in zip(actual, predicted))
    mean_square_error = sum_square_error / len(actual)
    return mean_square_error

print('평균제곱근오차', mse(ly_preds, y_test))
```

```python
plt.figure(figsize=(9,4))
plt.scatter(x_test['Whole_weight'], y_test, label='y_test')
plt.scatter(x_test['Whole_weight'], ly_preds, c='y', label='ly_preds')
#plt.plot(X\x_train, lr.predict(X_train), color='blue')
plt.show() 
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(x_train, y_train)

from sklearn.metrics import mean_squared_error

dy_preds = model.predict(x_test)
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x_test['Whole_weight'], y_test, label='y_test')
plt.scatter(x_test['Whole_weight'], dy_preds, c='g', label='dy_preds')
plt.show() 
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(x_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(x_test)
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
```

```python
plt.figure(figsize=(9,4))
plt.scatter(x_test['Whole_weight'], y_test, label='y_test')
plt.scatter(x_test['Whole_weight'], ry_preds, c='orange', label='ry_preds')

plt.show()  
```

```python
X_train, X_test, Y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD,Adam


model = Sequential()

model.add(Dense(16,input_shape=(8,),activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(2,activation='sigmoid'))

model.compile(optimizer='adam', loss="binary_crossentropy", metrics=['accuracy'])

model.summary()
```

```python
model_history = model.fit(x=X_train, y=Y_train, epochs=30, batch_size=32, validation_data=(X_test, y_test))
y_pred = model.predict(X_test)

y_test_class = np.argmax(y_test,axis=1)
y_pred_class = np.argmax(y_pred,axis=1)
```
